```python
purpose_list = [21030,"Rail transport","Industrial development","Power generation/non-renewable sources","RESCHEDULING AND REFINANCING"]
```

```python
data_r = df[["recipient","donor","coalesced_purpose_name","commitment_amount_usd_constant"]]
```

```python
data_r.coalesced_purpose_name = data_r.coalesced_purpose_name.str.lower()
```

```python
data_r
```

```python
data_r = data_r[data_r['coalesced_purpose_code'].isin([purpose.lower() for purpose in purpose_list])]
```

```python
data_r = data_r.groupby(['recipient','donor','coalesced_purpose_code'], as_index=False).agg({"commitment_amount_usd_constant": "sum"})
```

```python
data_r
```

```python
data_f = data_r.groupby(["recipient","donor"])['coalesced_purpose_name'].apply(list)
```

```python
data_f.to_csv("./final.csv")
```

```python
for purpose in purpose_list:
    data_r[data_r['coalesced_purpose_code'] == purpose].to_csv("./"+str(purpose)+".csv")
```
